The first thing we are going to is impute some values. We'll create two new columns, age_known and cabin_known, which will be true if we know the age and cabin respectively.

```python
df_test['known_age']= np.where(df_test.Age.isna(),False,True)
df_train['known_age']= np.where(df_train.Age.isna(),False,True)
df_test['known_cabin']= np.where(df_test.Cabin.isna(),False,True)
df_train['known_cabin']= np.where(df_train.Cabin.isna(),False,True)
```

We'll get back to age and cabin. We have the simple solution and we'll see if we can get more data out of it by making it more complex. Instead we'll look at the smaller sets of missing data. Let's start with the missing fair in the test data. We will have to impute this data as we can't drop the row per kaggle rules. Let's take a closer look at the individual.

```python
df_test[df_test.Fare.isnull()]
```

Mr. Thomas Storey. A third class passenger who embarked from Southampton. We have his approximate age but not his cabin. Given that this is the only passenger that we do not have a value for Fare I can be sure that this is MCAR (Missing Completely at Random) data. First thing to do is to see if anyone else was traveling on his ticket. It's unlikely but possible.

```python
df_test[df_test.Ticket=='3701']
```

That was a miss. To fill this I'm going to have to use deductive imputation and get the average value of a ticket from people that match Mr. Storey's details.

```python
df_merge=pd.concat([df_train,df_test],sort=False)
df_merge.shape
```

```python
df_merge[(df_merge.Pclass==3)&(df_merge.Embarked=='S')&(df_merge.Sex=='male')&(df_merge.SibSp==0&(df_merge.Parch==0))].Fare.mean()
```

Looking at the above the ticket probably cost &#0163;9.77 Let's impute that value into our data.

```python
df_test.Fare.fillna(9.77, inplace=True)
```

Now let's look at the two missing embark locations in the train dataset.

Looking at this we see two women that survived. They were traveling on the same ticket in first class and shared a Cabin. It would be nice to try and figure out where they probably Embarked from.

```python
df_merge.drop(['Age','Cabin'],axis =1,inplace=True)
```

```python
sns.countplot(x = 'Embarked', hue = 'Pclass',data = df_merge, orient='v')
```

The good news is that it's almost impossible that they departed from Queenstown. The bad news is that they could have come from Cherbourg or Southampton. I'm going against the advice of Han Solo and I'm going to have to ask for the odds.

Time to break out Bayesian statistics. My apologies to those who thought that we could get out of this without math. So if we took the three ports and randomly selected a person from there the odds per port are:

- $P(E_S)=\frac{1}{3}$
- $P(E_C)=\frac{1}{3}$
- $P(E_Q)=\frac{1}{3}$

That was easy. Now we need to figure out the odds that a person from first class is picked.

```python
df_merge.Embarked.value_counts()
df_merge[df_merge.Embarked=='S'].Pclass.value_counts()
df_merge[df_merge.Embarked=='C'].Pclass.value_counts()
df_merge[df_merge.Embarked=='Q'].Pclass.value_counts()
```
